```python
data_price=pd.read_csv("../blockchain data\external data\market-price.csv")
```

```python
datos_hash=pd.read_csv("../blockchain data\external data\hash-rate.csv")
```

```python
datos_hash["Timestamp"]=pd.to_datetime(datos_hash["Timestamp"])
```

```python
datos_hash= datos_hash.replace(0,np.nan)
```

```python
datos_hash=datos_hash.groupby(pd.Grouper(key="Timestamp", freq = '1D'))[["hash-rate"]].sum()
```

```python
datos_hash=datos_hash.interpolate()
```

```python
datos_hash
```

```python
data=pd.merge(datos_inner, datos_hash, how = 'left', on = 'Timestamp')
```

```python
data
```

```python
data=data[data["market-price"]!=0]
```

```python
data.shape
```

```python
data
```

```python
data["hash-rate"]= data["hash-rate"].replace(0,np.nan)
```

```python
data["hash-rate"]=data["hash-rate"].interpolate()
```

```python
data[data["Fee Reward"]==0]
```

```python
data["market-price"]=round(data["market-price"],2)
```

```python
data
```

```python
data=data.sort_values(by="Timestamp",ascending=False)
```

```python
data.to_csv("data_reward_casi.csv")
```

```python
fig, ax = plt.subplots()
ax.plot(data.index,data["hash-rate"])
fig.set_size_inches(18.5, 10.5)
ax.set(xlabel='Total Reward USD', ylabel='USD',
       title='History TH/s Value ')
ax.grid()
plt.show()
```

```python
data
```

```python
data["TH/USD Value"]=((data["Block Reward"]+data["Fee Reward"])/data["hash-rate"])*data["market-price"]
```
